```
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

```
img = cv2.imread("pizza_bluescreen.jpg")

img = cv2.cvtColor( img , cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

** Creating a mask **

```
img_copy = np.copy(img)

begin_range = np.array([0 , 0 , 220])
end_range = np.array([255 , 255 , 255])

mask = cv2.inRange( img_copy , begin_range , end_range )

plt.imshow(mask , cmap = "gray")
plt.show()
```

```
img_copy[ mask != 0 ] = [0 , 0 , 0]

plt.imshow(img_copy)
```

```
a , b , c = img_copy.shape

print(a , b , c)
```

```
back_ground = cv2.imread("space_background.jpg")

back_ground = cv2.cvtColor(back_ground , cv2.COLOR_BGR2RGB)

back_ground = cv2.resize(back_ground , (b , a))

plt.imshow(back_ground)
plt.show()
```

** Lets merge the both **

```
back_ground[ mask == 0] = [0 , 0 , 0]

plt.imshow(back_ground)
```

**  pizza for aliens **

```
final_img = back_ground + img_copy

plt.imshow(final_img)
```
