```
import numpy as np
import matplotlib.pyplot as plt 

import cv2
```

```
img = cv2.imread("water_balloons.jpg")

img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```
Red_channel = img[: , : , 0]
Green_channel = img[: , : , 1]
Blue_channel = img[: , : , 2]

f , (ax1 , ax2 , ax3) = plt.subplots(1 , 3 , figsize = (10 , 5))

ax1.set_title(" Red ")
ax1.imshow(Red_channel , cmap = "gray")

ax2.set_title(" Green ")
ax2.imshow(Green_channel , cmap = "gray")

ax3.set_title(" Blue ")
ax3.imshow(Blue_channel , cmap = "gray")
```

** Converting Image into HSV format **

```
img_new = cv2.imread("water_balloons.jpg")

hsv = cv2.cvtColor(img_new , cv2.COLOR_BGR2HSV)

plt.imshow(hsv)
plt.show()
```

```
Hue = hsv[: , : , 0]
Saturation = hsv[: , : , 1]
Value = hsv[: , : , 2]

f , (ax1 , ax2 , ax3) = plt.subplots(1 , 3 , figsize = (10 , 5))

ax1.set_title(" Hue ")
ax1.imshow(Hue , cmap = "gray")

ax2.set_title(" Saturation ")
ax2.imshow(Saturation , cmap = "gray")

ax3.set_title(" Value ")
ax3.imshow(Value , cmap = "gray")
```

** Hue seems good to bring all pink balloons out  **

```
img_copy = np.copy(hsv)

begin_range = np.array([160 , 0 , 0])
end_range = np.array([180 , 255 , 255])

hue_mask = cv2.inRange(img_copy , begin_range , end_range)

plt.imshow(hue_mask)
plt.show()
```

```
img[ hue_mask == 0 ] = [0 , 0 , 0]

plt.imshow(img)
plt.show()
```
