```
import numpy as np
import matplotlib.pyplot as plt 

import cv2 
```

```
img_solid = cv2.imread("pink_solid.jpg")

img_solid = cv2.cvtColor( img_solid , cv2.COLOR_BGR2RGB)

img_stripes = cv2.imread("stripes.jpg")

img_stripes = cv2.cvtColor( img_stripes , cv2.COLOR_BGR2RGB)

f , (ax1 , ax2) = plt.subplots( 1 , 2 ,figsize = (10 , 5))

ax1.set_title(" Solid Image ")
ax1.imshow( img_solid )

ax2.set_title(" Stripes Image ")
ax2.imshow( img_stripes)
```

```
Gray_solid = cv2.cvtColor( img_solid , cv2.COLOR_RGB2GRAY ) 
Gray_stripes = cv2.cvtColor( img_stripes , cv2.COLOR_RGB2GRAY ) 

norm_solid = Gray_solid / 255.0
norm_stripes = Gray_stripes / 255.0

def f_img( norm_img ):
    f = np.fft.fft2( norm_img )
    f_shift = np.fft.fftshift(f)
    freq = 50* np.log(np.abs(f_shift))
    
    return freq
```

```
f , (ax1 , ax2) = plt.subplots( 1 , 2 ,figsize = (10 , 5))

ax1.set_title(" Solid Image ")
ax1.imshow( img_solid )

ax2.set_title(" Transformed Solid Image ")
ax2.imshow(f_img(norm_solid) , cmap = "gray")
```

```
f , (ax1 , ax2) = plt.subplots( 1 , 2 ,figsize = (10 , 5))

ax1.set_title(" Stripes Image ")
ax1.imshow( img_stripes )

ax2.set_title(" Transformed Stripes Image ")
ax2.imshow(f_img(norm_stripes) , cmap = "gray")
```
