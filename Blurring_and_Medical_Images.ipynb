```
import numpy as np
import matplotlib.pyplot as plt 

import cv2
```

```
img = cv2.imread("brain_MR.jpg")

img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

** Lets see the difference between Blurred and Non Blurred Images **

** Without Blurring **

```
img_copy = np.copy(img)

img_copy = cv2.cvtColor(img_copy , cv2.COLOR_RGB2GRAY)

plt.imshow(img_copy , cmap = "gray")
plt.show()

sobel_X = np.array([[-1 , 0 ,  1],
                    [-2 , 0 ,  2], 
                    [-1 , 0 ,  1]])

sobel_Y = np.array([[-1 , -2 , -1],
                    [0 , 0 , 0],
                    [1 , 2 , 1]])

Horz_edges = cv2.filter2D(img_copy , -1 , sobel_X)
All_edges = cv2.filter2D( Horz_edges , -1 , sobel_Y ) 

plt.imshow(All_edges , cmap = "gray")
plt.show()
```

```
ret , thres_no_blur = cv2.threshold( All_edges , 50 , 255 , cv2.THRESH_BINARY)

plt.imshow(thres_no_blur , cmap = "gray")
plt.show()
```

** With Blurring **

```
img_copy = np.copy(img)

img_copy = cv2.cvtColor(img_copy , cv2.COLOR_RGB2GRAY)

img_blur = cv2.GaussianBlur( img_copy , (5,5) , 0)

Horz_edges = cv2.filter2D(img_blur , -1 , sobel_X)
All_edges = cv2.filter2D( Horz_edges , -1 , sobel_Y ) 

ret , thres_no_blur = cv2.threshold( All_edges , 50 , 255 , cv2.THRESH_BINARY)

plt.imshow(thres_no_blur , cmap = "gray")
plt.show()
```

** You could see after smoothing , there's less noise in the Image and edges appear more clear  **
