```
import numpy as np
import matplotlib.pyplot as plt 

import cv2
```

```
img = cv2.imread("sunflower.jpg")

img = cv2.cvtColor( img , cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```
img_copy = np.copy(img)

img_copy = cv2.cvtColor(img_copy , cv2.COLOR_RGB2GRAY)

plt.imshow(img_copy , cmap = "gray")
plt.show()
```

```
wide_range_canny = cv2.Canny( img_copy , 0 , 100 )
narrow_range_canny = cv2.Canny( img_copy , 180 , 240 )

f , (ax1 , ax2) = plt.subplots( 1 , 2 , figsize = (10 , 5))

ax1.set_title(" Wide Range ")
ax1.imshow( wide_range_canny , cmap = "gray")

ax2.set_title(" Narrow Range ")
ax2.imshow( narrow_range_canny , cmap = "gray")
```
