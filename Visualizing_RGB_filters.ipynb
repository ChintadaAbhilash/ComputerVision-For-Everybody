```
import numpy as np
import matplotlib.pyplot as plt
```

```
import cv2
```

```
img = cv2.imread("Hermionedhface.jpg")

img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```
Red_filter = img[: , : , 0]
Blue_filter = img[: , : , 1]
Green_filter = img[: , : , 2]

f , (ax1 , ax2 , ax3) = plt.subplots(1 , 3 , figsize = (10,5))

ax1.set_title(" Red Channel ")
ax1.imshow(Red_filter , cmap = "gray")

ax2.set_title(" Blue Channel ")
ax2.imshow(Blue_filter , cmap = "gray")


ax3.set_title(" Green Channel ")
ax3.imshow(Green_filter , cmap = "gray")
```
