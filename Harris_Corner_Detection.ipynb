```
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

```
img = cv2.imread("waffle.jpg")

img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```
img_copy = np.copy(img)

img_gray = cv2.cvtColor(img_copy , cv2.COLOR_RGB2GRAY)

img_gray = img_gray.astype(np.float32)

plt.imshow(img_gray , cmap = "gray")
plt.show()
```

```
Corners = cv2.cornerHarris( img_gray , 2 , 3 , 0.04)

plt.imshow(Corners , cmap = "gray")
```

corners look very dull , lets dialate to make them visible

```
Corners = cv2.dilate( Corners , None )

plt.imshow(Corners , cmap = "gray")
```

Extracting and Displaying Strong corners 

```
threshold = 0.1 * Corners.max()

print(threshold)
```

```
Corners_img = np.copy(img)

for i in range(0 , Corners.shape[0]):
    for j in range(0 , Corners.shape[1]):
        if(Corners[i][j] > threshold):
            cv2.circle(Corners_img , (j , i) , 1  , (0 , 255 , 0) , 1)
            
plt.imshow(Corners_img)
plt.show()
```
